```python
missing_value_in_percentage=(cup.isnull().sum()/cup.shape[0])*100
missing_value_in_percentage
```

#Filtering coloums which have more than 80% percentage of null values 

```python
def more_columns(missing_value):
  cols = []
  for key, value in missing_value.items():
    if value > 80:
      cols.append(key)
  return cols
more_columns(missing_value_in_percentage)
# Even if the coloums contains more than 80 percentage null value we cannot drop because of the nature of the data
```

# Question 3: Number of  yellow and red card awarded to argentina in different world cup

```python
cup['away_red_card'].unique()
```

### Here the nan value means 0 card awarded so we will replace nan value with Zero(0) and the values with player name to One(1) for better visualzation

```python
# For red cards
cup['home_red_card'] = cup['home_red_card'].fillna(0)  
cup['home_red_card'] = cup['home_red_card'].apply(lambda x: 0 if x == 0 else 1)  
cup['away_red_card'] = cup['away_red_card'].fillna(0)  
cup['away_red_card'] = cup['away_red_card'].apply(lambda x: 0 if x == 0 else 1)  
# For yellow carda
cup['home_yellow_card_long'] = cup['home_yellow_card_long'].fillna(0)  
cup['home_yellow_card_long'] = cup['home_yellow_card_long'].apply(lambda x: 0 if x == 0 else 1)  
cup['away_yellow_card_long'] = cup['away_yellow_card_long'].fillna(0)  
cup['away_yellow_card_long'] = cup['away_yellow_card_long'].apply(lambda x: 0 if x == 0 else 1)  
```

### Ploting graph for visualization

```python
# Making small dataframe for the precised result and with organized coloumns name
cards_by_year = cup.groupby('Year').apply(lambda x: pd.Series({
    'red_cards': x['home_red_card'][x['home_team'] == 'Argentina'].sum() + x['away_red_card'][x['away_team'] == 'Argentina'].sum(),
    'yellow_cards': x['home_yellow_card_long'][x['home_team'] == 'Argentina'].sum() + x['away_yellow_card_long'][x['home_team'] == 'Argentina'].sum()
})).reset_index()
cards_by_year


fig, axes = plt.subplots(1, 2, figsize=(12, 6))  
sns.barplot(data=cards_by_year, x='Year', y='red_cards', color='red', label='Red Cards', ax=axes[0])
axes[0].set_title('Red Cards by Year')  
sns.barplot(data=cards_by_year, x='Year', y='yellow_cards', color='yellow', label='Yellow Cards', ax=axes[1])
axes[1].set_title('Yellow Cards by Year') 
axes[0].legend()
axes[1].legend()
plt.tight_layout()
plt.show()
```

The charts represent the number of red and yellow cards Argentina received in different FIFA World Cup tournaments over the years.

Red Cards Interpretation (Left Chart)
Argentina received red cards in 1982 (2 red cards) and 1990 (1 red card).

No red cards were recorded in other tournaments.

This suggests Argentina generally avoided direct expulsions in most World Cups.

Yellow Cards Interpretation (Right Chart)
Argentina received a high number of yellow cards in 1978 and 1986 (10 yellow cards each).

Other notable years with many yellow cards include 1998, 2010, 2014, and 2022.

The lowest number of yellow cards was in 2018, with only 2.

This indicates fluctuating disciplinary records across different tournaments.

Overall, Argentina has had more yellow cards than red cards, showing they engaged in physical play but managed to avoid excessive dismissals in most tournaments.

# Question no 5: Win loss and draws of argentina in different worldcup, those matches which went in penalty is not counted

```python
cup['result'] = cup.apply(lambda row: 'Win' if (row['home_team'] == 'Argentina' and row['home_score'] > row['away_score']) or \
                                                                     (row['away_team'] == 'Argentina' and row['away_score'] > row['home_score']) else \
                                                                     'Draw' if row['home_score'] == row['away_score'] else \
                                                                     'Loss', axis=1)

results_by_year = cup.groupby(['Year', 'result'])['result'].count().reset_index(name='count')

print(results_by_year)
```

```python
years = results_by_year['Year'].unique()
fig, axes = plt.subplots(1, num_years, figsize=(60, 6))  
for i, year in enumerate(years):
    year_data = results_by_year[results_by_year['Year'] == year]
    axes[i].pie(year_data['count'], labels=year_data['result'], autopct='%1.1f%%', startangle=90)
    axes[i].set_title(f'Argentina Match Results in {year}')
plt.tight_layout()
plt.show()
```

The pie chart shows the win,loss and draw record of argentina in different world cup excluding the matches which went on penalty

# Question no 7: Is there relation between average attendence and goal scored in argentina games during world cup
